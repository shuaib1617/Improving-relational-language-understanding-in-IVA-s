```python
def check_data_quality(dataset, dataset_name):
    # Check for missing values
    missing_values = dataset.isnull().sum()

    # Check for duplicates
    duplicate_rows = dataset.duplicated().sum()

    # Check for outliers (considering only numerical columns)
    numerical_columns = dataset.select_dtypes(include=[np.number]).columns
    outliers = dataset[numerical_columns].apply(lambda x: (x - x.mean()).abs() > 3 * x.std()).sum()

    # Check for data consistency
    consistency_checks = []

    # Additional consistency checks based on dataset columns
    binary_columns = ['MultiIntent', 'Greeting', 'Backstory', 'Justification', 'Rant', 'Gratitude', 'Other', 'Express Emotion']
    for col in binary_columns:
        if col in dataset.columns:
            unique_values = dataset[col].unique()
            if len(unique_values) != 2 or set(unique_values) - {0, 1}:
                consistency_checks.append(f"{col} should be binary (0 or 1): {unique_values}")

    # Results
    print(f"Checking data quality for {dataset_name} dataset:")
    print("\nMissing Values:")
    print(missing_values[missing_values > 0])

    print("\nDuplicate Rows:")
    print(duplicate_rows)

    print("\nOutliers:")
    print(outliers[outliers > 0])

    print("\nData Consistency Checks:")
    for check in consistency_checks:
        print(check)

# Load datasets
x_y_align_data = pd.read_csv('x_y_align.csv')
all_data_df = pd.read_csv('all_data_by_threshold.csv')
tagged_selections_df = pd.read_csv('tagged_selections_by_sentence.csv')
all_multi_intent_df = pd.read_csv('all_multi_intent.csv')

# Calling the function
check_data_quality(x_y_align_data, "x_y_align")
check_data_quality(all_data_df, "All Data")
check_data_quality(tagged_selections_df, "Tagged Selections")
check_data_quality(all_multi_intent_df, "All Multi-Intent")
```

```python
def preprocess_dataset(dataset, dataset_name):
    # Create binary indicators for missing values in numeric columns
    numeric_columns = dataset.select_dtypes(include=[np.number]).columns
    for col in numeric_columns:
        dataset[f'{col}_missing'] = pd.isnull(dataset[col]).astype(int)

        # Solve for outliers (replace with mean for simplicity)
        mean_value = dataset[col].mean()
        dataset[col] = np.where((dataset[col] - mean_value).abs() > 3 * dataset[col].std(), mean_value, dataset[col])

    # Create binary indicators for missing values in categorical columns
    categorical_columns = dataset.select_dtypes(include=[object]).columns
    consistency_checks = []  # Initialize consistency_checks
    for col in categorical_columns:
        dataset[f'{col}_missing'] = pd.isnull(dataset[col]).astype(int)

        # Solve for anomalies (replace with mode for simplicity)
        if col == 'MultiIntent':
            dataset[col] = np.where((dataset[col] != 0) & (dataset[col] != 1), dataset[col].mode()[0], dataset[col])
            consistency_checks.append(f"MultiIntent should be binary (0 or 1): {dataset['MultiIntent'].unique()}")

    # Display results after preprocessing
    print(f"Preprocessing for {dataset_name} dataset:")
    print("\nMissing Value Indicators:")
    print(dataset.filter(regex='_missing'))

    print("\nOutliers (replaced with mean):")
    print(dataset[numeric_columns].apply(lambda x: (x - x.mean()).abs() > 3 * x.std()).sum())

    print("\nAnomalies (replaced with mode):")
    for check in consistency_checks:
        print(check)
    
    return dataset

# Apply the function to your datasets
preprocessed_all_data = preprocess_dataset(all_data_df, "All Data")
preprocessed_tagged_selections = preprocess_dataset(tagged_selections_df, "Tagged Selections")
preprocessed_all_multi_intent = preprocess_dataset(all_multi_intent_df, "All Multi-Intent")
```

```python
# Define the check_data_quality function
def check_data_quality(data):
    # Check for missing values
    missing_values = data.isnull().sum()
    print("Missing values:\n", missing_values)

    # Check for duplicate rows
    duplicate_rows = data.duplicated()
    print("\nNumber of duplicate rows:", duplicate_rows.sum())

    # Check for outliers in numerical columns
    numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
    outliers = pd.DataFrame()
    for column in numerical_columns:
        q1 = data[column].quantile(0.25)
        q3 = data[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers[column] = (data[column] < lower_bound) | (data[column] > upper_bound)

    print("\nOutliers in numerical columns:\n", outliers.sum())

    # Check for anomalies in categorical columns
    categorical_columns = data.select_dtypes(include=['object']).columns
    anomalies = pd.DataFrame()
    for column in categorical_columns:
        anomalies[column] = data[column].str.contains('\[.*\]', regex=True)

    print("\nAnomalies in categorical columns:\n", anomalies.sum())

# Apply the function to preprocessed_all_data
print("Checking quality of preprocessed_all_data:")
check_data_quality(preprocessed_all_data)

# Apply the function to preprocessed_tagged_selections
print("\nChecking quality of preprocessed_tagged_selections:")
check_data_quality(preprocessed_tagged_selections)
```

```python
import pandas as pd

def merge_datasets(x_y_align, tagged_selections_by_sentence, all_data_by_threshold, all_multi_intent):
    """
    Merge multiple datasets step by step using common columns.

    Parameters:
    - x_y_align: DataFrame
    - tagged_selections_by_sentence: DataFrame
    - all_data_by_threshold: DataFrame
    - all_multi_intent: DataFrame

    Returns:
    - merged_df: DataFrame, the merged dataset
    """

    # Merge x_y_align and tagged_selections_by_sentence DataFrames using common column 'Sentence ID'
    merged_df = pd.merge(x_y_align, tagged_selections_by_sentence, how='inner', left_on='Sentence ID', right_on='SentenceID')

    # Optionally, drop the duplicate 'SentenceID' column after merging
    merged_df.drop('SentenceID', axis=1, inplace=True)

    # Merge with all_data_by_threshold using common columns 'Unselected' and 'Selected'
    merged_df = pd.merge(merged_df, all_data_by_threshold, how='inner', on=['Unselected', 'Selected'])

    # Merge with all_multi_intent using common column 'Dataset ID'
    merged_df = pd.merge(merged_df, all_multi_intent, how='inner', on='Dataset ID')

    return merged_df

# Example usage:
# merged_data = merge_datasets(x_y_align, tagged_selections_by_sentence, all_data_by_threshold, all_multi_intent)
```
